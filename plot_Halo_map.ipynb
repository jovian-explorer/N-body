```python
import numpy as np
import proplot as plot
```

```python
f = open('/home/idies/workspace/Storage/midsuman/AA_609/data/Halo_map_1.000') #path to map
N = np.fromfile(f, count=3, dtype='uint64')
N1,N2,N3 = N
print(N1,N2,N3)
l = np.fromfile(f, count=1, dtype='float32')
print(l)
data = np.fromfile(f, count=N1*N2*N3, dtype='float32')
f.close()
data = np.reshape(data, (N1,N2,N3), order='C')
```

```python
index = 16
image = data[0:N1,0:N2,index]
fig, ax = plot.subplots(figsize=(3,3.5))
vmax =np.max(image)
print(vmax)
levels = [0,1,10,1e2,1000]
im = ax.imshow(image, interpolation='gaussian', origin='lower', levels=levels, cmap='inferno')
ax.colorbar(im, loc='t', label='$log(1+\delta_{halo})$', formatter='log')
ax.format(grid=False, xlabel='Grids', ylabel='Grids')
filename='Halo_map.png'
plot.Figure.savefig(fig,filename)
```
