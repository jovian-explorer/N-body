```python
import numpy as np
import proplot as plot
```

```python
f = open('/home/idies/workspace/Storage/Avinash28799/persistent/density/Halo_map_8.000') #path to map
N  = np.fromfile(f, count=3, dtype='uint64')
N1,N2,N3 = N
print(N1,N2,N3)
l = np.fromfile(f, count=1, dtype='float32')
print(l)
data = np.fromfile(f, count=N1*N2*N3, dtype='float32')
f.close()
data = np.reshape(data, (N1,N2,N3), order='C')
```

```python
index = 16 #x-y slice
image = data[0:N1,0:N2,index]
fig, ax = plot.subplots(figsize=(3,3.5))
#levels = [0,1,10,1e2,1e3,1e4,1e5]
im = ax.imshow(image, interpolation='gaussian', origin='lower', cmap='turbo')
ax.colorbar(im, loc='t', label='$log(1+\delta)$')
ax.format(grid=False, xlabel='Grids', ylabel='Grids')
filename= 'DM_map.png'
plot.Figure.savefig(fig,'8.jpg')
```
